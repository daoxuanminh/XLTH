# Lượng tử 64 mức (6 bit) - dùng phương pháp truncation


```python
N = 200
n = np.arange (N)
f0 = 1/50
delta = 2/64
xn = np.sin(2 * math.pi * f0 * n)
x = (xn/delta)
j=0
xqn = arr.array ('d',n)
for i in x:
    i = int(i)
    xqn[j] = i*delta
    j=j+1
#print(xqn)
```

```python
#plt.plot (n,xqn)
plt.title ("Đồ thị xqn truncation")
plt.xlabel ("Trục n",fontsize=16)
plt.ylabel ("Trục giá trị xqn",fontsize=16)
plt.stem(n,xqn,'-g')
plt.show ()

#plt.plot (n,xqn)
plt.title ("Đồ thị en truncation")
en = xn - xqn
plt.xlabel ("Trục n",fontsize=16)
plt.ylabel ("Trục giá trị en",fontsize=16)
#plt.stem(n,en,'-g')
plt.plot (n,en)
plt.show ()

a = sum(en**2)
Pq = a/200
#print (Pq)
y = 0.5/Pq
SQNR = 10*math.log10(y)
print ("SQNR<truncation> = ",SQNR)

N = 200
n = np.arange (N)
f0 = 1/50
xn = np.sin(2 * math.pi * f0 * n)
x = (xn/delta)
j=0
xqn = arr.array ('d',n)
for i in x:
    i = round (i)
    xqn[j] = i*delta
    j=j+1

#plt.plot (n,xqn)
plt.title ("Đồ thị xqn round")
plt.xlabel ("Trục n",fontsize=16)
plt.ylabel ("Trục giá trị xqn",fontsize=16)
plt.stem(n,xqn,'-g')
plt.show ()

plt.title ("Đồ thị en round")
en = xn - xqn
plt.xlabel ("Trục n",fontsize=16)
plt.ylabel ("Trục giá trị en",fontsize=16)
#plt.stem(n,en,'-g')
plt.plot (n,en)
plt.show ()

a = sum(en**2)
Pq = a/200
#print (Pq)
y = 0.5/Pq
SQNR = 10*math.log10(y)
print ("SQNR<round> = ",SQNR)
```
