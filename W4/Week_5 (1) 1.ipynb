# Đào Xuân Minh - 20205100

```python
import wave 
import numpy as np
import matplotlib.pyplot as plt 
import math
```

```python

file = wave.open('Thoai_008.wav', 'r')
signal = file.readframes(-1)
signal= np.frombuffer(signal, dtype = "int16") # chuyển về kiểu int với độ dài 16bit
signal = np.array(signal, dtype='float64')
Fs = file.getframerate() #Fs = 44100
#print(len(signal))
N_10ms = int(Fs*0.01) # số mẫu trong 10ms = 441
N_100ms = int(Fs*0.1) # số mẫu trong 100ms là 4410
#print(N_signal)
time_signal = np.linspace( 0 , len(signal)/Fs , num = len(signal) )
so_mau_nang_luong = int (len(signal)/0.01/Fs)+1
#print (so_mau_nang_luong)
P = np.arange(so_mau_nang_luong,dtype='int64')
j=0
Time = np.arange(so_mau_nang_luong,dtype='float64')
#print (Time)
for i in range(0,len(signal),N_10ms):
    P[j] = sum(signal[i:i+N_100ms]**2)
    j=j+1
for i in range(so_mau_nang_luong):
    Time[i] = 0.01*i
#print (Time)
#print((P))
```

```python
plt.figure(figsize=(12,6))
plt.plot(time_signal, signal,color = "blue")
plt.title("Sound Wave")
plt.xlabel("Time(s)")
plt.show()
```

# Tín hiệu năng lượng

```python
P_max = max(P)
so_mu = -1
thuong = P_max;
while (thuong != 0):
    thuong = int(thuong/10)
    so_mu = so_mu+1
print (so_mu)
plt.figure(figsize=(12,6))
y = 0.05*10**so_mu
plt.plot([0,Time[-1]],[y,y] ,color = "red")
plt.plot(Time, P,color = "blue")
plt.title("Năng lượng")
plt.xlabel("Time(s)")
plt.show()
```

Phóng to đoạn thời gian 0,5s đến 2s

```python
plt.figure(figsize=(12,6))
plt.plot(Time[50:200], P[50:200],color = "blue")
plt.title("Năng lượng")
plt.xlabel("Time(s)")
plt.show()
```

# Xác định thời gian bắt đầu và kết thúc của tín hiệu âm thanh

```python

start=0
for i in P:
    start=start+1
    if (i >= 0.05*10**so_mu):
        break
time_start = start*0.01
print('Thời gian bắt đầu là: ' +str(time_start)+ 's')

end = start
j=start
for i in P[start:-1]:
    j = j+1
    if (i>=0.05*10**so_mu):
        end = j
time_end = end*0.01
print('Thời gian kết thúc là: ' +str(time_end)+ 's')
```
